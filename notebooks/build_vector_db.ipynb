```python
import json
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
filepath_input = '/content/drive/MyDrive/MayakBot/mayakovsky-corpus.jsonl'
filepath_output = '/content/drive/MyDrive/MayakBot/chroma_db_mayakovsky'
```

```python
documents = []
with open(filepath_input, 'r', encoding='utf-8') as f:
    for line in f:
        chunk = json.loads(line)
        doc = Document(
            page_content=chunk['text'],
            metadata=chunk['metadata']
        )
        documents.append(doc)

print(f"Загружено {len(documents)} чанков.")
```

```python
print("Скачиваем модель BGE-M3...")
embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3",
    model_kwargs={'device': 'cuda'},
    encode_kwargs={'batch_size': 16})
```

```python
print("Создаем векторную базу и считаем векторы...")
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=embeddings,
    persist_directory=filepath_output
)

print(f"Готово! Папка {filepath_output} создана.")
```
